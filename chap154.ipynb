```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('view1.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
kp,des = sift.detectAndCompute(gray,None)
img2=cv2.drawKeypoints(gray,kp,img) 
plt.imshow(img)
plt.show()

plt.imshow(img2)
plt.show()
```
