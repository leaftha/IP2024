```python
import cv2
import numpy as np
img = cv2.imread("sIMG_2164.jpg")
# img = cv2.imread('dave.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
lines = cv2.HoughLines(edges,1,np.pi/180,200)
for rho,theta in lines[0]:
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
    cv2.imwrite('houghlines3.jpg',img)
plt.imshow(img)
```

```python
import cv2
import numpy as np
img = cv2.imread('sIMG_2164.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
minLineLength = 100
maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength,maxLineGap)
for x1,y1,x2,y2 in lines[0]:
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
cv2.imwrite('sIMG_2164.jpg',img)
plt.imshow(img)
```

```python
import cv2
import numpy as np
def nothing(x):
    pass
img = cv2.imread('sIMG_2164.jpg')
min = 100
max = 10
linenums = 10
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize = 3)
lines = cv2.HoughLinesP(edges,1,np.pi/180,100,min,max)
# Create a black image, a window
cv2.namedWindow('image')
# create trackbars for color change
cv2.createTrackbar('min','image',0,500,nothing)
cv2.createTrackbar('max','image',0,500,nothing)
cv2.createTrackbar('linenums','image',0,100,nothing)
cv2.setTrackbarPos('min','image',min)
cv2.setTrackbarPos('max','image',max)
cv2.setTrackbarPos('linenums','image',linenums)
while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
     break
    min = cv2.getTrackbarPos('min','image')
    max = cv2.getTrackbarPos('max','image')
    linenums = cv2.getTrackbarPos('linenums','image')
    lines = cv2.HoughLinesP(edges,1,np.pi/180,100,min,max)
    linenum = min(linenums, lines.shape[0]) 
    for i in range(linenum):
       for x1,y1,x2,y2 in lines[i]:
        cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
cv2.destroyAllWindows()
```
