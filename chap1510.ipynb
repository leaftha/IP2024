```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
img1 = cv2.imread('acat.png',0) # queryImage
img2 = cv2.imread('acat2.png',0) # trainImage
# Initiate SIFT detector
sift = cv2.SIFT_create()
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
# BFMatcher with default params
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
marcher = cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
matches = marcher.match(des1,des2)
# Apply ratio test
matches = sorted(matches,key = lambda x:x.distance)
# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:200],None,flags=cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
plt.imshow(img3),plt.show()
```
